```python
import torch
import numpy as np
import pandas as pd

# Load datasets
spectra_data = pd.read_csv("/home/sym/natural_gas/plots _for_paper/standard_simulated_spectra_interp_excess.csv").values  # shape (10000, 600)
concentration_data = pd.read_csv("/home/sym/natural_gas/plots _for_paper/standard_simulated_concentrations_interp_excess.csv").values  # shape (10000, 7)
# 
# Reshape spectra for CNN-style input: (Nobs, 1, 600)
dataset_clean = np.reshape(spectra_data, (spectra_data.shape[0], 1, spectra_data.shape[1]))

# Convert to torch tensors
Spec_T = torch.tensor(dataset_clean, dtype=torch.float32)
Con_T = torch.tensor(concentration_data, dtype=torch.float32)

# Define training params
Nobs = dataset_clean.shape[0]
batch_size = 16
device = torch.device("cpu")  # forced to CPU due to CUDA issues

Spec_T = Spec_T.to(device)
Con_T = Con_T.to(device)

# Split into train, val, test (80/10/10 split)
train_features = Spec_T[:int(Nobs * 0.8)]
val_features = Spec_T[int(Nobs * 0.8):int(Nobs * 0.9)]
test_features = Spec_T[int(Nobs * 0.9):]

train_targets = Con_T[:int(Nobs * 0.8)]
val_targets = Con_T[int(Nobs * 0.8):int(Nobs * 0.9)]
test_targets = Con_T[int(Nobs * 0.9):]

# Create PyTorch Datasets and DataLoaders
train = torch.utils.data.TensorDataset(train_features, train_targets)
val = torch.utils.data.TensorDataset(val_features, val_targets)
test = torch.utils.data.TensorDataset(test_features, test_targets)

train_loader = torch.utils.data.DataLoader(train, batch_size=batch_size, shuffle=True)
valid_loader = torch.utils.data.DataLoader(val, batch_size=batch_size)
test_loader = torch.utils.data.DataLoader(test, batch_size=batch_size)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np

class CNNRegressor(nn.Module):
    def __init__(self):
        super(CNNRegressor, self).__init__()
        self.conv1 = nn.Conv1d(in_channels=1, out_channels=32, kernel_size=3)
        self.batch_norm1 = nn.BatchNorm1d(32)
        self.relu = nn.ReLU()
        self.pool = nn.MaxPool1d(kernel_size=2)
        self.conv2 = nn.Conv1d(in_channels=32, out_channels=64, kernel_size=3)
        self.batch_norm2 = nn.BatchNorm1d(64)
        self.flatten = nn.Flatten()
        self.fc1 = nn.Linear(9472, 128)
        self.batch_norm3 = nn.BatchNorm1d(128)
        self.fc2 = nn.Linear(128, 7)

    def forward(self, x):
        x = self.conv1(x)
        x = self.batch_norm1(x)
        x = self.relu(x)
        x = self.pool(x)
        x = self.conv2(x)
        x = self.batch_norm2(x)
        x = self.relu(x)
        x = self.pool(x)
        x = self.flatten(x)
        x = self.fc1(x)
        x = self.batch_norm3(x)
        x = self.relu(x)
        x = self.fc2(x)
        return x


# Assuming you have appropriate data loaders for training and testing
# train_loader and test_loader should be instances of DataLoader class

# Initialize the model
model = CNNRegressor().to(device)

# Define loss function and optimizer for regression
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.0001)

# Training the model
num_epochs = 18
train_losses1 = []

for epoch in range(num_epochs):
    model.train()
    epoch_loss = 0.0

    for inputs, targets in train_loader:
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, targets)
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()

    avg_epoch_loss = epoch_loss / len(train_loader)
    train_losses1.append(avg_epoch_loss)
    print(f"Epoch {epoch+1}/{num_epochs}, Loss: {loss.item()}")

# Evaluate the model on the test set

model.eval()

with torch.no_grad():
    total_loss = 0
    total_samples = 0
    for inputs, targets in test_loader:
        outputs = model(inputs)
        loss = criterion(outputs, targets)
        total_loss += loss.item() * targets.size(0)
        total_samples += targets.size(0)

    average_loss = total_loss / total_samples
    print(f"Test Average Loss: {average_loss}")
```

```python
import torch
import numpy as np
import pandas as pd

# Load datasets
spectra_data = pd.read_csv("/home/sym/natural_gas/plots _for_paper/standard_simulated_spectra_interp_C5+.csv").values  # shape (10000, 600)
concentration_data = pd.read_csv("/home/sym/natural_gas/plots _for_paper/standard_simulated_concentrations_interp_C5+.csv").values  # shape (10000, 7)
# 
# Reshape spectra for CNN-style input: (Nobs, 1, 600)
dataset_clean = np.reshape(spectra_data, (spectra_data.shape[0], 1, spectra_data.shape[1]))

# Convert to torch tensors
Spec_T = torch.tensor(dataset_clean, dtype=torch.float32)
Con_T = torch.tensor(concentration_data, dtype=torch.float32)

# Define training params
Nobs = dataset_clean.shape[0]
batch_size = 16
device = torch.device("cpu")  # forced to CPU due to CUDA issues

Spec_T = Spec_T.to(device)
Con_T = Con_T.to(device)

# Split into train, val, test (80/10/10 split)
train_features = Spec_T[:int(Nobs * 0.8)]
val_features = Spec_T[int(Nobs * 0.8):int(Nobs * 0.9)]
test_features = Spec_T[int(Nobs * 0.9):]

train_targets = Con_T[:int(Nobs * 0.8)]
val_targets = Con_T[int(Nobs * 0.8):int(Nobs * 0.9)]
test_targets = Con_T[int(Nobs * 0.9):]

# Create PyTorch Datasets and DataLoaders
train = torch.utils.data.TensorDataset(train_features, train_targets)
val = torch.utils.data.TensorDataset(val_features, val_targets)
test = torch.utils.data.TensorDataset(test_features, test_targets)

train_loader = torch.utils.data.DataLoader(train, batch_size=batch_size, shuffle=True)
valid_loader = torch.utils.data.DataLoader(val, batch_size=batch_size)
test_loader = torch.utils.data.DataLoader(test, batch_size=batch_size)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np

class CNNRegressor(nn.Module):
    def __init__(self):
        super(CNNRegressor, self).__init__()
        self.conv1 = nn.Conv1d(in_channels=1, out_channels=32, kernel_size=3)
        self.batch_norm1 = nn.BatchNorm1d(32)
        self.relu = nn.ReLU()
        self.pool = nn.MaxPool1d(kernel_size=2)
        self.conv2 = nn.Conv1d(in_channels=32, out_channels=64, kernel_size=3)
        self.batch_norm2 = nn.BatchNorm1d(64)
        self.flatten = nn.Flatten()
        self.fc1 = nn.Linear(9472, 128)
        self.batch_norm3 = nn.BatchNorm1d(128)
        self.fc2 = nn.Linear(128, 7)

    def forward(self, x):
        x = self.conv1(x)
        x = self.batch_norm1(x)
        x = self.relu(x)
        x = self.pool(x)
        x = self.conv2(x)
        x = self.batch_norm2(x)
        x = self.relu(x)
        x = self.pool(x)
        x = self.flatten(x)
        x = self.fc1(x)
        x = self.batch_norm3(x)
        x = self.relu(x)
        x = self.fc2(x)
        return x


# Assuming you have appropriate data loaders for training and testing
# train_loader and test_loader should be instances of DataLoader class

# Initialize the model
model = CNNRegressor().to(device)

# Define loss function and optimizer for regression
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.0001)

# Training the model
num_epochs = 20
train_losses1 = []

for epoch in range(num_epochs):
    model.train()
    epoch_loss = 0.0

    for inputs, targets in train_loader:
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, targets)
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()

    avg_epoch_loss = epoch_loss / len(train_loader)
    train_losses1.append(avg_epoch_loss)
    print(f"Epoch {epoch+1}/{num_epochs}, Loss: {loss.item()}")

# Evaluate the model on the test set

model.eval()

with torch.no_grad():
    total_loss = 0
    total_samples = 0
    for inputs, targets in test_loader:
        outputs = model(inputs)
        loss = criterion(outputs, targets)
        total_loss += loss.item() * targets.size(0)
        total_samples += targets.size(0)

    average_loss = total_loss / total_samples
    print(f"Test Average Loss: {average_loss}")
```

```python
# Full clean code to load, compute, filter by wavenumber range, and plot cross-sections (m^2/mol) for last two species

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the CSV file
file_path = "PNNL_CS.csv"
df = pd.read_csv(file_path)

# Constants
T = 22 + 273.15  # Temperature in Kelvin
R = 8.314e-5     # Ideal gas constant in m^3·bar/(mol·K)
P_bar = 1.01325  # Pressure in bar
L_m = 1.0        # Path length in meters
X = 1e-6         # Mole fraction for 1 ppm

# Determine the number of spectra
num_columns = df.shape[1]
num_spectra = num_columns // 2

# Extract last two spectra (2*2 columns)
last_two_spectra = df.iloc[:, -4:]

# Prepare plot
# Prepare the plot with thicker lines, no grid, and proper labels
plt.figure(figsize=(10, 6))

# Storage for average computation
all_cross_sections = []

labels = ['n-Hexane', 'n-Heptane']

for idx, i in enumerate(range(0, 4, 2)):
    wn = last_two_spectra.iloc[:, i].values
    ab = last_two_spectra.iloc[:, i + 1].values

    # Filter valid entries
    valid = ~(pd.isna(wn) | pd.isna(ab))
    wn = wn[valid].astype(float)
    ab = ab[valid].astype(float)

    # Restrict to wavenumber range 2963–2968 cm^-1
    range_mask = (wn >= 2963) & (wn <= 2968)
    wn = wn[range_mask]
    ab = ab[range_mask]

    # Calculate cross-section
    abs_ln = ab * np.log(10)
    cross_sec_m2_per_mol = (R * T) * abs_ln / (X * P_bar * L_m)
    all_cross_sections.append(cross_sec_m2_per_mol)

    plt.plot(wn, cross_sec_m2_per_mol, linewidth=2.5, label=labels[idx])

# Compute and plot average cross-section
avg_cross_section = np.mean(np.array(all_cross_sections), axis=0)
plt.plot(wn, avg_cross_section, '--', linewidth=3, label='Average', color='white')

# Final plot settings
plt.xlabel('Wavenumber (cm$^{-1}$)', fontsize=18)
plt.ylabel('Cross-section (m$^2$/mol)', fontsize=18)
# plt.title('Absorption Cross-section (2963–2968 cm$^{-1}$)', fontsize=14)
plt.legend(fontsize = 25)
plt.tight_layout()
plt.savefig("/home/sym/natural_gas/plots _for_paper/CS_C5+.pdf", transparent=True, dpi=300)
plt.savefig("/home/sym/natural_gas/plots _for_paper/CS_C5+.png", transparent=True, dpi=300)
plt.show()
plt.show()
```

```python
# Full combined clean code

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

# Step 1: Load absorbance data
file_path = "PNNL_CS.csv"
df = pd.read_csv(file_path)

# Step 2: Constants
ln10 = np.log(10)
L_new = 0.10  # 10 cm path length

# Step 3: Define gas species and extraction order
gas_names_ordered = ["Methane", "Ethane", "Propane", "nC4", "iC4", "nC5", "iC5", "nC6", "nC7"]
num_species = len(gas_names_ordered)

# Step 4: Extract and store spectra
gas_spectra = {}
for i in range(num_species):
    wn = df.iloc[:, 2 * i].astype(float).values
    ab = df.iloc[:, 2 * i + 1].astype(float).values
    gas_spectra[gas_names_ordered[i]] = (wn, ab)

# Step 5: Define mole fractions
mole_fractions = {
    "Methane": 0.5612 / 1e-6,
    "Ethane": 0.011 / 1e-6,
    "Propane": 0.0058 / 1e-6,
    "nC4": 0.005 / 1e-6,
    "iC4": 0.0030 / 1e-6,
    "nC5": 0.0020 / 1e-6,
    "iC5": 0.000 / 1e-6,
    "nC6": 0.04 * 1e-2 / 1e-6,
    "nC7": 0.42 * 1e-2 / 1e-6,
}

# Step 6: Adjust absorbance using mole fractions
adjusted_spectra = {}
for gas in gas_names_ordered:
    wn, ab = gas_spectra[gas]
    x = mole_fractions[gas]
    ab_new = ab * x * L_new * ln10
    adjusted_spectra[gas] = (wn, ab_new)

# Step 7: Interpolation setup
interpolated_wn = np.linspace(2963, 2968, 600)
absorbance_zoomed_interp = []
for gas in gas_names_ordered:
    wn, ab = adjusted_spectra[gas]
    mask = (wn >= 2963) & (wn <= 2968)
    interp_func = interp1d(wn[mask], ab[mask], kind="linear", fill_value="extrapolate")
    absorbance_zoomed_interp.append(interp_func(interpolated_wn))
absorbance_zoomed_interp = np.stack(absorbance_zoomed_interp)  # (9, 600)

# Step 8: Generate two synthetic mixtures
np.random.seed(0)
mixture_no_C6C7 = np.random.rand(9)
mixture_no_C6C7[7:] = 0  # Set nC6 and nC7 contributions to zero
abs_no_C6C7 = mixture_no_C6C7 @ absorbance_zoomed_interp

mixture_with_C6C7 = np.random.rand(9)
abs_with_C6C7 = mixture_with_C6C7 @ absorbance_zoomed_interp

# Step 9: Compute residual
residual = abs_with_C6C7 - abs_no_C6C7

# Step 10: Plot everything
plt.figure(figsize=(10, 6))
plt.plot(interpolated_wn, abs_no_C6C7, label=r"$A_{\mathrm{C1\text{--}C5}}$ Predicted by Cascade Model", linewidth=2)
plt.plot(interpolated_wn, abs_with_C6C7, label="Total", linewidth=2, linestyle='--')
plt.plot(interpolated_wn, residual, label="Residual", linewidth=2.5, linestyle=':')
plt.xlabel("Wavenumber (cm$^{-1}$)", fontsize=12)
plt.ylabel("Absorbance", fontsize=12)
# plt.title("Synthetic Absorbance Spectra and Residual", fontsize=14)
plt.legend(fontsize = 18)
plt.tight_layout()
plt.style.use('dark_background')
plt.savefig("/home/sym/natural_gas/plots _for_paper/predvstrue_C5+.pdf", transparent=True, dpi=300)
plt.savefig("/home/sym/natural_gas/plots _for_paper/predvstrue_C5+.png", transparent=True, dpi=300)
plt.show()
plt.show()

plt.show()
```
